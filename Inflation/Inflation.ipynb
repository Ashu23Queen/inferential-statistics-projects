### Inflation

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
```

```python
# 1. Loading dataset

data = {
    'Inflation': [2.5, 2.1, 3.2, 5.0, 8.0, 7.5, 6.5, 3.5, 4.0, 3.0],
    'Interest_Rate': [1.25, 1.5, 2.0, 1.0, 3.5, 4.5, 5.0, 4.5, 4.0, 3.5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# 2. Descriptive Statistics
print("--- Descriptive Statistics ---")
print(df.describe())
```

```python
# 3. Correlation Analysis (Pearson's r)
correlation, p_value = stats.pearsonr(df['Interest_Rate'], df['Inflation'])
```

```python
print(f"\nCorrelation Coefficient: {correlation:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 4. Simple Linear Regression (Inferential)
X = df['Interest_Rate']
y = df['Inflation']
X = sm.add_constant(X) 
```

```python
model = sm.OLS(y, X).fit()
print("\n--- Regression Summary ---")
print(model.summary())
```

```python
# 5. Visualizing the Inference
plt.figure(figsize=(10, 6))
sns.regplot(x='Interest_Rate', y='Inflation', data=df, ci=95) # ci=95 adds the confidence interval
plt.title('Regression Analysis: Interest Rates vs. Inflation')
plt.xlabel('Interest Rate (%)')
plt.ylabel('Inflation Rate (%)')
plt.grid(True)
plt.show()
```
