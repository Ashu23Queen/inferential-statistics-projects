```python
import numpy as np
import scipy.stats as stats
```

```python
# 1. Sample Data for Two Individual Groups 
 
grp1_snooze_counts = np.array([
    3, 4, 1, 5, 2, 3, 6, 2, 4, 3, 
    1, 5, 2, 3, 4, 1, 5, 2, 3, 4,
    1, 5, 2, 3, 4, 1, 5, 2, 3, 4
])
grp1_snooze_counts
```

```python
# Group 2: E.g., Working Professionals (n=35)
grp2_snooze_counts = np.array([
    1, 2, 0, 3, 1, 4, 2, 1, 0, 3,
    5, 1, 2, 2, 0, 1, 3, 4, 1, 2,
    0, 1, 1, 2, 3, 1, 0, 2, 1, 3,
    5, 1, 2, 4, 2
])
grp2_snooze_counts
```

```python
# 2. Descriptive Statistics 
mean1 = np.mean(grp1_snooze_counts)
mean2 = np.mean(grp2_snooze_counts)
n1 = len(grp1_snooze_counts)
n2 = len(grp2_snooze_counts)
```

```python
print(" ---  Descriptive statistics --- ")
print(f"Group 1 Mean (Students): {mean1: .3f} (n = {n1}) ")
print(f"Group 2 Mean (Professionals): {mean2:.3f} (n={n2})\n")
```
